## Segment 2 - Defining elements of a plot

```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

### Defining axes, ticks, and grids

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
# add axes
ax = fig.add_axes([.1,.1,1,1])

# plot
ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

# add limites to axes
ax.set_xlim([1,9])
ax.set_ylim([0,5])

# tick marks
ax.set_xticks([0,2,3,4,5,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

# add limites to axes
ax.set_xlim([1,9])
ax.set_ylim([0,5])

# tick marks
ax.set_xticks([0,2,3,4,5,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

# add grid to background of chart
ax.grid()
ax.plot(x,y)
```

### Generating multiple plots in one figure with subplots 

```python
fiig = plt.figure()
# create tuple
fig, (ax1, ax2) = plt.subplots(1,2)

# 1 figure with 2 axes, 2 plots within 1 figure
ax1.plot(x)
ax2.plot(x,y)
```
